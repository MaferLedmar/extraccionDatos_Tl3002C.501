## FILTROS BÁSICOS

```python
#filtro por objeto
filtro1 = df[df["NOMBRE"] == "INDUSTRIAS LANCERMEX"]
filtro1
```

```python
#filtro por filas
filtro2 = df.iloc[3:10, :] # Filas consecutivas
filtro3 = df.iloc[ [8,24,100] , : ] #filas no consecutivas

#significa que solo quiero las filas 8,24, y 100, pero todas las columnas
filtro3
```

```python
#Filtro por columnas
filtro4 = df.iloc[ : , 1:3] #Consecuticas
filtro5 = df.iloc[ : , [1,4,5]] #No consecutivas

filtro5
```

##**FILTROS BÁSICOS PARA VARIABLES CUANTITATIVAS**

```python
filtro8 =df[df["MONTO ADEUDADO"] > 70000]
filtro8
```

##FILTRO POR CONECTORES

```python
#filtro y
filtro9 = df[(df['MONTO ADEUDADO'] < 6000) & (df["FECHA_FACTURA"] == "2023-03-14")]
filtro9
#7
```

```python
#filtro o (or)
filtro10 = df[(df['MONTO ADEUDADO'] < 6000) | (df["FECHA_FACTURA"] == "2023-03-14")]
filtro10
```

```python
#filtro not
filtro11 = df[~(df['MONTO ADEUDADO'] >= 6000) & ~(df["FECHA_FACTURA"] == "2023-03-14")]
filtro11
```
